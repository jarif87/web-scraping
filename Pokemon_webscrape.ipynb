```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url="https://pokemondb.net/pokedex/all"
```

```python
r=requests.get(url)
r
```

```python
soup=BeautifulSoup(r.text,"lxml")
# soup
```

```python
table=soup.find("table",class_="data-table sticky-header block-wide")
title=table.find_all("th")
header=[]
for i in title:
  name=i.text
  header.append(name)
print(header)
```

```python
df=pd.DataFrame(columns=header)
df
```

```python
row=table.find_all("tr")
for i in row[1:]:

  data=i.find_all("td")
  rows=[tr.text for tr in data]
  l=len(df)
  df.loc[l]=rows
```

```python
df
```

```python
df.drop(columns="#",index=1,inplace=True)
```

```python
df.to_csv("project_pokemon.csv",index=False)
```
