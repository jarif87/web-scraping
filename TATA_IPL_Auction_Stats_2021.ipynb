```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url="https://www.iplt20.com/auction/2021"
```

```python
r=requests.get(url)
r
```

```python
soup=BeautifulSoup(r.text,"lxml")
```

```python
table=soup.find("table",class_="ih-td-tab auction-tbl")
title=table.find_all("th")
header=[]
for i in title:
  name=i.text
  header.append(name)
print(header)
```

```python
df=pd.DataFrame(columns=header)
```

```python
row=table.find_all("tr")
for i in row[1:]:

  data=i.find_all("td")
  rows=[tr.text for tr in data]
  l=len(df)
  df.loc[l]=rows
```

```python
df.to_csv("my_first.csv",index=False)
```
