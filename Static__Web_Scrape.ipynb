```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url="https://en.wikipedia.org/wiki/Red_states_and_blue_states"
```

```python
r=requests.get(url)
r
```

```python
soup=BeautifulSoup(r.text,"lxml")
# soup
```

```python
table = soup.find_all("table", class_="wikitable")
title = table[-1].find_all("tr")
header = [i.text.strip() for i in title[0] if i.text.strip()]

print(header)
```

```python
df=pd.DataFrame(columns=header)
df
```

```python
# Assuming table is a ResultSet obtained from soup.find_all("table", class_="wikitable")

for t in table:
    rows = t.find_all("tr")[1:]  # Skip the header row
    for row in rows:
        data = row.find_all("td")
        row_values = [td.text.strip() for td in data]

        # Check if the length of row_values matches the number of columns in the DataFrame
        if len(row_values) == len(df.columns):
            l = len(df)
            df.loc[l] = row_values
        else:
            print(f"Ignoring row {row_values} as it has mismatched columns.")
```
