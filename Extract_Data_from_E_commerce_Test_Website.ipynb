```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets"
```

```python
r=requests.get(url)
print(r)
```

```python
soup=BeautifulSoup(r.text,"lxml")
# soup
```

```python
boxes=soup.find_all("div",class_="col-md-4 col-xl-4 col-lg-4")
```

```python
len(boxes)
```

```python
names=soup.find_all("a",class_="title")
for i in names:
  print(i.text)
```

```python
import pandas as pd
data = [{"Title": i.text} for i in names]

# Create a DataFrame from the list of dictionaries
df = pd.DataFrame(data)
df
```

```python
prices=soup.find_all("h4",class_="float-end price card-title pull-right")
```

```python
for m in prices:
  print(m.text)
```

```python
data = []
for price in prices:
    data.append({"Price": price.text})
```

```python
df0=pd.DataFrame(data)
df0
```

```python
result_df = pd.concat([df,df0],axis=1)
result_df
```

```python
result_df.to_csv("new_df.csv",index=False)
```
